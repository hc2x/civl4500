## 1 Creating and Manipulating Arrays

First import numpy and matplotlib

````python
import numpy as np
import matplotlib.pyplot as plt
````

### 1.1. Create two 2D arrays representing coordinates x, y on the cartesian plane
Both should cover the range (-2, 2) and have 100 points in each direction

````python
# Create two 2D arrays representing coordinates x, y on the cartesian plane
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)

# Reshape x and y into 2D arrays using meshgrid
x, y = np.meshgrid(x, y)
````

### 1.2. Visualize each 2D array using `pcolormesh`

Use the correct coordiantes for the x and y axes.

````python
# Define a figure
plt.figure(figsize=(10, 5))
# Plotting x coordinates
plt.subplot(1, 2, 1)
plt.pcolormesh(x, y, x)
plt.xlabel("x")
plt.ylabel("y")
plt.title("2D Array of x coordinates")
plt.colorbar()
# Plotting y coordinates
plt.subplot(1, 2, 2)
plt.pcolormesh(x, y, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("2D Array of y coordinates")
plt.colorbar()
# Automatic-adjusting and showing the figure
plt.tight_layout()
plt.show()
````

### 1.3 From your cartesian coordinates, create polar coordinates $r$ and $\varphi$

Refer to the [wikipedia page](https://en.wikipedia.org/wiki/Polar_coordinate_system#Converting_between_polar_and_Cartesian_coordinates) for the conversion formula. You will need to use numpy's `arctan2` function. Read its documentation.


---

![image](https://upload.wikimedia.org/wikipedia/commons/thumb/7/78/Polar_to_cartesian.svg/440px-Polar_to_cartesian.svg.png)

The polar coordinates r and φ can be converted to the Cartesian coordinates x and y by using the trigonometric functions sine and cosine:

   $$x = r \cos \varphi$$

   $$y = r \sin \varphi$$

The Cartesian coordinates x and y can be converted to polar coordinates r and φ with r ≥ 0 and φ in the interval (−π, π] by:

   $$r = \sqrt{x^2 + y^2}$$

   $$\varphi = \text{atan2}(y, x)$$

````python
# Conversion to polar coordinates
r = np.sqrt(x**2 + y**2)
phi = np.arctan2(y, x)
````

### 1.4. Visualize $r$ and $\varphi$ on the 2D  $x$ / $y$ plane using `pcolormesh`

````python
# Define a figure
plt.figure(figsize=(10, 5))
# Plotting r coordinates
plt.subplot(1, 2, 1)
plt.pcolormesh(x, y, r)
plt.xlabel("x")
plt.ylabel("y")
plt.title("2D Array of r coordinates")
plt.colorbar()
# Plotting phi coordinates
plt.subplot(1, 2, 2)
plt.pcolormesh(x, y, phi)
plt.xlabel("x")
plt.ylabel("y")
plt.title("2D Array of phi coordinates")
plt.colorbar()
# Automatic-adjusting and showing the figure
plt.tight_layout()
plt.show()
````

### 1.5 Caclulate the quanity $f = \cos^2(4r) + \sin^2(4\varphi)$

And plot it on the x$ / $y$ plane

````python
# Calculate the quantity f
f = np.cos(4 * r)**2 + np.sin(4 * phi)**2
print(f.shape)

# Visualize f on the x / y plane
plt.figure(figsize=(6, 6))  # Adjust figure size if needed
plt.pcolormesh(x, y, f)
plt.xlabel("x")
plt.ylabel("y")
plt.title("2D Array of f = cos^2(4r) + sin^2(4φ)")
plt.colorbar()
plt.tight_layout()
plt.show()
````

### 1.6 Plot the mean of f with respect to the x axis

as a function of y

````python
# Calculate the mean of f with respect to the x-axis
mean_f_x = np.mean(f, axis=0)
print(mean_f_x)
# Plot the mean of f as a function of y
plt.figure(figsize=(6, 4))
plt.plot(x[0, :], mean_f_x)  # Use y[0, :] for the y-coordinates
plt.xlabel("y")
plt.ylabel("Mean of f")
plt.title("Mean of f with respect to x-axis")
plt.grid(True)
plt.tight_layout()
plt.show()
````

### 1.7 Plot the mean of f with respect to the y axis

as a function of x

````python
# Calculate the mean of f with respect to the y-axis
mean_f_y = np.mean(f, axis=1)
print(mean_f_y)
# Plot the mean of f as a function of x
plt.figure(figsize=(6, 4))
plt.plot(y[:, 0], mean_f_y)  # Use x[:, 0] for the x-coordinates
plt.xlabel("x")
plt.ylabel("Mean of f")
plt.title("Mean of f with respect to y-axis")
plt.grid(True)
plt.tight_layout()
plt.show()
````

### 1.8 Plot the mean of $f$ with respect to $\phi$ as a function of $r$

This is hard. You will need to define a discrete range of $r$ values and then figure out how to average $f$ within the bins defined by your $r$ grid. There are many different ways to accomplish this.

````python
# Define a discrete range of r values (bins)
r_bins = np.linspace(r.min(), r.max(), 20)  # Adjust the number of bins as needed

# Initialize an array to store the mean of f for each r bin
mean_f_phi = np.zeros_like(r_bins)

# Iterate through the r bins
for i in range(len(r_bins) - 1):
    # Find indices where r falls within the current bin
    indices = np.where((r >= r_bins[i]) & (r < r_bins[i + 1]))

    # Calculate the mean of f for those indices
    if len(indices[0]) > 0:  # Check if there are any points within the bin
        mean_f_phi[i] = np.mean(f[indices])
    else:
        mean_f_phi[i] = np.nan  # Or another appropriate value for empty bins


# Plot the mean of f as a function of r
r_bin_centers = (r_bins[:-1] + r_bins[1:]) / 2 # Calculate the centers of the bins
plt.figure(figsize=(8, 6))
plt.plot(r_bin_centers, mean_f_phi[:-1])
plt.xlabel("r")
plt.ylabel("Mean of f")
plt.title("Mean of f with respect to φ as a function of r")
plt.grid(True)
plt.tight_layout()
plt.show()
````

## Part 2: Analyze [ARGO](http://www.argo.ucsd.edu) Data

In this problem, we use real data from ocean profiling floats.
ARGO floats are autonomous robotic instruments that collect Temperature, Salinity, and Pressure data from the ocean. ARGO floats collect one "profile" (a set of messurements at different depths or "levels").

<img src="https://argo.ucsd.edu/wp-content/uploads/sites/361/2020/06/float_cycle_1.png" width="80%" />

Each profile has a single latitude, longitude, and date associated with it, in addition to many different levels.

Let's start by using [pooch](https://www.fatiando.org/pooch/latest/) to download the data files we need for this exercise.
The following code will give you a list of `.npy` files that you can open in the next step.

````python
!pip install pooch
import pooch
url = "https://www.ldeo.columbia.edu/~rpa/float_data_4901412.zip"
files = pooch.retrieve(url, processor=pooch.Unzip(), known_hash="2a703c720302c682f1662181d329c9f22f9f10e1539dc2d6082160a469165009")
files
````

### 2.1 Load each data file as a numpy array.

You can use whatever names you want for your arrays, but I recommend

`T`: temperature

`S`: salinity

`P`: pressure

`date`: date

`lat`: latitude

`lon`: longitude

`level`: depth level

**Note**: you have to actually look at the file name (the items in `files`) to know which files corresponds to which variable.


````python
# Load each data file as a numpy array
level = np.load(files[0])
lat = np.load(files[1])
T = np.load(files[2])
P = np.load(files[3])
S = np.load(files[4])
lon = np.load(files[5])
date = np.load(files[6])
````

### 2.2 Examine the shapes of T, S and P compared to `lon`, `lat`, `date` and `level`. How do you think they are related?

Based on the shapes, which dimensions do you think are shared among the arrays?

````python
# Examine the shapes of T, S and P compared to lon, lat, date, and level.
print("Shape of T:", T.shape)
print("Shape of S:", S.shape)
print("Shape of P:", P.shape)
print("Shape of date:", date.shape)
print("Shape of lat:", lat.shape)
print("Shape of lon:", lon.shape)
print("Shape of level:", level.shape)
````

- The shapes of T, S, and P are (78, 79), indicating 78 profiles with 79 depth levels each.
- The shapes of lon, lat, and date are (78,), indicating 78 profiles.
- The shape of level is (79,), indicating 79 depth levels.

This suggests that each profile (indexed by lon, lat, and date) contains measurements at 79 different depth levels (indexed by level).


### 2.3 Make a plot for each column of data in T, S and P (three plots).

The vertical scale should be the `levels` data. Each plot should have a line for each column of data. It will look messy.

````python
# Plot each column of data in T
plt.figure(figsize=(10, 6))
for i in range(T.shape[1]):
    plt.plot(T[:,i], level, label=f'Profile {i+1}')
plt.gca().invert_yaxis()
plt.xlabel('Temperature (°C)')
plt.ylabel('Depth (m)')
plt.title('Temperature Profiles')
plt.show()

# Plot each column of data in S
plt.figure(figsize=(10, 6))
for i in range(S.shape[1]):
    plt.plot(S[:,i], level, label=f'Profile {i+1}')
plt.gca().invert_yaxis()
plt.xlabel('Salinity (PSU)')
plt.ylabel('Depth (m)')
plt.title('Salinity Profiles')
plt.show()

# Plot each column of data in P
plt.figure(figsize=(10, 6))
for i in range(P.shape[1]):
    plt.plot(P[:,i], level, label=f'Profile {i+1}')
plt.gca().invert_yaxis()
plt.xlabel('Pressure (dbar)')
plt.ylabel('Depth (m)')
plt.title('Pressure Profiles')
plt.show()
````

### 2.4 Compute the mean and standard deviation of each of T, S and P at each depth in `level`.

````python
mean_T = np.mean(T, axis=1)
std_T = np.std(T, axis=1)
mean_S = np.mean(S, axis=1)
std_S = np.std(S, axis=1)
mean_P = np.mean(P, axis=1)
std_P = np.std(P, axis=1)
````

### 2.5 Now make three similar plot, but show only the mean T, S and P at each depth. Show [error bars](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.errorbar.html) on each plot using the standard deviations.


````python
# Temperature
plt.figure(figsize=(10, 6))
plt.plot(mean_T, level, label='Mean Temperature')
plt.fill_betweenx(level, mean_T - std_T, mean_T + std_T, alpha=0.3, label='Standard Deviation')
plt.gca().invert_yaxis()
plt.xlabel('Temperature (°C)')
plt.ylabel('Depth (m)')
plt.title('Mean and Standard Deviation of Temperature')
plt.legend()
plt.show()

# Salinity
plt.figure(figsize=(10, 6))
plt.plot(mean_S, level, label='Mean Salinity')
plt.fill_betweenx(level, mean_S - std_S, mean_S + std_S, alpha=0.3, label='Standard Deviation')
plt.gca().invert_yaxis()
plt.xlabel('Salinity (PSU)')
plt.ylabel('Depth (m)')
plt.title('Mean and Standard Deviation of Salinity')
plt.legend()
plt.show()

# Pressure
plt.figure(figsize=(10, 6))
plt.plot(mean_P, level, label='Mean Pressure')
plt.fill_betweenx(level, mean_P - std_P, mean_P + std_P, alpha=0.3, label='Standard Deviation')
plt.gca().invert_yaxis()
plt.xlabel('Pressure (dbar)')
plt.ylabel('Depth (m)')
plt.title('Mean and Standard Deviation of Pressure')
plt.legend()
plt.show()

````

### 2.6 Account For Missing Data

The profiles contain many missing values. These are indicated by the special "Not a Number" value, or `np.nan`.

When you take the mean or standard deviation of data with NaNs in it, the entire result becomes NaN. Instead, if you use the special functions `np.nanmean` and `np.nanstd`, you tell NumPy to ignore the NaNs.

Recalculate the means and standard deviations as in the previous sections using these functions and plot the results.

````python
# Use below code block for 2.4
mean_T = np.nanmean(T, axis=1)
std_T = np.nanstd(T, axis=1)
mean_S = np.nanmean(S, axis=1)
std_S = np.nanstd(S, axis=1)
mean_P = np.nanmean(P, axis=1)
std_P = np.nanstd(P, axis=1)
````
